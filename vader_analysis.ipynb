```python
# Pandas and NLTK Modules
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk.tokenize import sent_tokenize, word_tokenize, RegexpTokenizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('stopwords')
nltk.download('vader_lexicon')
nltk.download('punkt')

# Plotly Libraries
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objs as go
```

```python
df = pd.read_csv('localnews.csv')
#df = df.drop(['Link', 'Index'], axis=1)
df['Body'] = df['Body'].astype(str)
df.info()
```

```python
df['Body'] = df['Body'].apply(lambda txt: txt.lower())
df.head()
```

```python
stop_words=stopwords.words('english')
df['Body'] = df['Body'].apply(lambda txt: ' '.join([word for word in txt.split() if word not in stop_words]))
df.head()
```

```python
df['Body'] = df['Body'].apply(lambda txt: sent_tokenize(txt))
df.head()
```

```python
df['Body'] = df['Body'].apply(lambda txt: ' '.join(txt))
```

```python
#df.head()
#df.to_csv('body_tokenized.csv')
```

```python
sid = SentimentIntensityAnalyzer()
df['score'] = df['Body'].apply(lambda txt: sid.polarity_scores(txt))
df['negative'] = df['score'].apply(lambda txt: txt['neg'])
df['neutral'] = df['score'].apply(lambda txt: txt['neu'])
df['positive'] = df['score'].apply(lambda txt: txt['pos'])
df['compound'] = df['score'].apply(lambda txt: txt['compound'])

df.head()
```

```python
def polarity_score(compound):
    if compound > 0.05:
        return "positive"
    elif compound < -0.05:
        return "negative"
    elif compound >= -0.05 and compound < 0.05:
        return "neutral"

df['sentiment'] = df['compound'].apply(lambda val: polarity_score(val))
df.head(25)
```

```python
df['sentiment'].value_counts()
```

```python
fig = px.pie(df, names='sentiment', 
             height=300, width=600, 
             hole=0.7, 
             title='Sentiment Overview, VADER',
             color_discrete_sequence=['#CAE7B9', '#EB9486', '#F3DE8A'])
fig.update_layout(plot_bgcolor='#fafafa', paper_bgcolor='#fafafa')
fig.show()
```
