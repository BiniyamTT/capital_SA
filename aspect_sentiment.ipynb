```python
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objs as go
```

```python
df = pd.read_csv('final_with_body.csv')
df.info()
```

```python
df['Body'] = df['Body'].apply(str.lower)
df.head()
```

```python
def search(sstr):
    searched_df = df[df['Body'].str.contains(sstr.lower())]
    df_aspects = pd.read_csv('aspects.csv')
    df_positive = searched_df[searched_df['label']=='positive']
    df_negative = searched_df[searched_df['label']=='negative']
    df_neutral = searched_df[searched_df['label']=='neutral']
    new = {'aspect':sstr, 'positive':len(df_positive), 'negative':len(df_negative), 'neutral':len(df_neutral)}
    df_toappend = pd.DataFrame(data=new, index=[1])
    df_aspects= pd.concat([df_aspects, df_toappend], ignore_index=True)
    df_aspects.to_csv('aspects.csv', index=False)
```

```python
search('Education')
search('Health')
search('Industry')
search('Employment')
search('Economy')
search('Tourism')
search('Climate')
search('Infrastructure')
search('Law')
```
